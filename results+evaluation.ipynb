```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```
t_timee = {
    'time (min)' : [12.8,9.2,6.4],
    'no. of rows' : [1000,700,500],

}
time_textrank = pd.DataFrame(t_timee)
fig = sns.lineplot(data=time_textrank, x = "no. of rows", y = "time (min)",  color='red')
fig.set_title("TextRank Time Taken")
plt.show()
```

```
b_timee = {
    'time (min)' : [240,166,97],
    'no. of rows' : [1000,700,500],

}
time_bert = pd.DataFrame(b_timee)
fig = sns.lineplot(data=time_bert, x = "no. of rows", y = "time (min)", color='green')
fig.set_title("BERT Time Taken")
plt.show()
```

```
#Time taken for BERT and TextRank to compute

compare_time = {
    'no. of rows' : [1000,700,500],
    'textrank_time (min)' : [12.8,9.2,6.4],
    'bert_time (min)' : [240,166,97]
}

text_bert_compare_time = pd.DataFrame(compare_time)


fig, ax = plt.subplots(1,2,sharex=True, figsize=(10,5))
sns.lineplot(ax = ax[0], x = "no. of rows", y = "textrank_time (min)", color='r', data = text_bert_compare_time)
ax[0].set_title("TextRank Time")
ax[0].set(ylabel="time")
sns.lineplot(ax = ax[1], x = "no. of rows", y = "bert_time (min)",color='g', data = text_bert_compare_time)
ax[1].set_title("Bert Time")
ax[1].set(ylabel="time")
plt.show()

```

```
#comparing textrank with and without stopwords

stopwords_comapre = {
    'metric'  : ['Precision','Recall','F-Measure'],
    'kept'    : [45,44,44],
    'removed' : [32,17,22]
}

stopwords_eval = pd.DataFrame(stopwords_comapre)

fig, ax = plt.subplots(1,2,sharex=True, figsize=(10,5))
sns.barplot(ax = ax[0], x = "metric", y = "kept", palette='summer', data = stopwords_eval)
ax[0].set_title("Removed Stopwords")
ax[0].set(ylabel="Percentage")
sns.barplot(ax = ax[1], x = "metric", y = "removed",palette='spring', data = stopwords_eval)
ax[1].set_title("Kept Stopwords")
ax[1].set(ylabel="Percentage")
plt.show()

```

```
rouge_scores = {
    'metrics' : ['Precision','Recall','F-Measure'],
    'textrank' : [46,44,44],
    'bert' : [56,22,32]
}

text_bert_compare_scores = pd.DataFrame(rouge_scores)

fig, ax = plt.subplots(1,2,sharex=True, figsize=(10,5))
sns.barplot(ax = ax[0], x = "metrics", y = "textrank", palette='autumn', data = text_bert_compare_scores)
ax[0].set_title("TextRank Evaluation")
ax[0].set(ylabel="Percentage")
sns.barplot(ax = ax[1], x = "metrics", y = "bert",palette='winter', data = text_bert_compare_scores)
ax[1].set_title("BERT Evaluation")
ax[1].set(ylabel="Percentage")

plt.show()
```
